```python
import random
```

```python

UserWins = 0
UserLoss = 0
CompWins = 0
CompLoss = 0
Tie = 0
```

```python
def Userwin(UserWins, CompLoss):
    print('\n Computer lost 1 point')
    CompLoss = CompLoss + 1
    print(' User earned 1 point')
    UserWins = UserWins + 1
    print(" Total earned points by User : ", UserWins)
    return UserWins, CompLoss
```

```python
def Compwin(UserLoss, CompWins):
    print('\n User lost 1 point')
    UserLoss = UserLoss + 1
    print(' Computer earned 1 point')
    CompWins = CompWins + 1
    print(" Total earned points by Computer : ", CompWins)
    return UserLoss, CompWins
```

```python
print("\t\t __________________________________________________________________________________")
print('\n\t\t  ************************Welcome to ROCK-PAPER-SCISSOR Game**********************')
print("\t\t __________________________________________________________________________________")
print("\n\n\t\t\t\t\t\t Let's get started....")
print("\t\t ----------------------------------------------------------------------------------")


while(1):
    print('\n\n 1.R for Rock \n 2.P for Paper \n 3.S for Scissor \n 4.E for Exit')
    UserInput = input('\n Enter your choice:')
    
    if(UserInput == 'R' or UserInput == 'P' or UserInput == 'S' or UserInput == 'E'):
        pass
    
    else:
        print(' Wrong choice!')
        continue

    CompChoice = ['R','P','S']
    CompInput = str(random.choice(CompChoice))     
        
        
    if(UserInput == CompInput):
        print(" It is Tie")
        Tie = Tie + 1
        print(' Total no of ties: ', Tie)
        
                
    elif(UserInput == 'E'):
        
        if(UserWins == CompWins):
            print('\n\n\n --------------------------------------------------------------------------------------------------------------')
            print('\n\n \t\t\t\t\t It is a Tie!!!!  Well played')
            
        elif(UserWins > CompWins):
            print('\n\n\n --------------------------------------------------------------------------------------------------------------')
            print('\n\n \t\t\t     */*/*/*/*/*/*/* Congratulations!! *\*\*\*\*\*\*\*')
            print("\t\t\t\t\t\t You win!!")
            
        else:
            print('\n\n\n --------------------------------------------------------------------------------------------------------------')
            print("\n\n \t\t\t\t\t\t You lose!!")
            print('\t\t\t\t\t   Better try next time!!')
            
        print("\n\n\n\n ************************************************SCOREBOARD**************************************************")
        print('\n ______________________________________________________________________________________________________________')
        print('\n\t\t User \t\t\t\t\t\t\t Computer')
        print(' ______________________________________________________________________________________________________________')
        print('\n\t\t Losses: ', UserLoss, ' \t\t\t\t\t\t Losses: ', CompLoss)
        print('\t\t Wins: ', UserWins, ' \t\t\t\t\t\t Wins: ', CompWins)
        print('\t\t\t\t\t\t Total Ties: ', Tie)
        print('--------------------------------------------------------------------------------------------------------------')
        break;
        
                
    elif(UserInput != CompInput):
        if(UserInput == 'R' and CompInput == 'S'):
            UserWins, CompLoss = Userwin(UserWins, CompLoss)

        elif(UserInput == 'R' and CompInput == 'P'):
            UserLoss, CompWins = Compwin(UserLoss, CompWins)

        elif(UserInput == 'S' and CompInput == 'R'):
            UserLoss, CompWins = Compwin(UserLoss, CompWins)

        elif(UserInput == 'P' and CompInput == 'R'):
            UserWins, CompLoss = Userwin(UserWins, CompLoss)

        elif(UserInput == 'P' and CompInput == 'S'):
            UserLoss, CompWins = Compwin(UserLoss, CompWins)

        elif(UserInput == 'S' and CompInput == 'P'):
            UserWins, CompLoss = Userwin(UserWins, CompLoss)
```
